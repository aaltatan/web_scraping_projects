```python
import pandas as pd
import numpy as np
import json
```

```python
with open("companies.jsonl", "r", encoding="utf-8") as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.DataFrame(data)
```

```python
df.shape
```

```python
fb = df['social_media'].apply(lambda lst: [x for x in lst if 'facebook' in x])
tw = df['social_media'].apply(lambda lst: [x for x in lst if 'twitter' in x])
insta = df['social_media'].apply(lambda lst: [x for x in lst if 'insta' in x])
linkedin = df['social_media'].apply(lambda lst: [x for x in lst if 'linkedin' in x])
yt = df['social_media'].apply(lambda lst: [x for x in lst if 'youtube' in x])
```

```python
df = pd.concat([df, fb, tw, insta, linkedin, yt], axis='columns')
```

```python
df.columns = ['title', 
              'category', 
              'posted_at', 
              'description', 
              'social_media',
              'address', 
              'website', 
              'phone', 
              'facebook', 
              'twitter',
              'instagram', 
              'linkedin', 
              'youtube']
```

```python
df['facebook'] = df['facebook'].apply(pd.Series).iloc[:, :1]
df['twitter'] = df['twitter'].apply(pd.Series).iloc[:, :1]
df['instagram'] = df['instagram'].apply(pd.Series).iloc[:, :1]
df['linkedin'] = df['linkedin'].apply(pd.Series).iloc[:, :1]
df['youtube'] = df['youtube'].apply(pd.Series).iloc[:, :1]
```

```python
df = df.drop(['social_media'], axis='columns')
```
