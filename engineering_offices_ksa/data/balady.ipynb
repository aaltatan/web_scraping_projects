```python
import pandas as pd
import json
```

```python
with open('balady.jsonl', 'r', encoding='utf-8') as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.DataFrame(data)
```

```python
df[['المنطقة', 'المدينة']] = (
    df['المنطقة \\ المدينة']
    .str
    .split(r'\\', regex=True, expand=True)
)
```
