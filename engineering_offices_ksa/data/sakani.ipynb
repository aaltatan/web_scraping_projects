```python
import pandas as pd
import json
```

```python
with open('sakani.jsonl', 'r', encoding='utf-8') as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.json_normalize(data)
```

```python
df = df[[
    'attributes.cr_or_lic_number',
    'attributes.classification_grade',
    'attributes.facility_name',
    'attributes.facility_size',
    'attributes.region',
    'attributes.city',
    'attributes.contact_number',
    'attributes.meter_price',
    'attributes.cr_number',
    'attributes.membership',
    'attributes.membership_status',
    'attributes.license_number',
    'attributes.logo',
    'attributes.logo_ext',
    'attributes.email',
    'attributes.supervision_max_price',
    'attributes.supervision_min_price',
    'attributes.turkey_max_price',
    'attributes.turkey_min_price',
    'attributes.design_contractor_max_price',
    'attributes.design_contractor_min_price'
]]
```

```python
df.dropna(thresh=len(df)*0.9, axis='columns', inplace=True)
```
