```python
import pandas as pd
import numpy as np
import json
from glob import glob
```

```python
data: list[dict] = []
for f in glob("*.jsonl"):
  with open(f, 'r', encoding='utf-8') as file:
    file_data = (json.loads(line[:-1]) for line in file.readlines())
    data += file_data
```

```python
df = pd.json_normalize(data)
```

```python
df.shape
```

```python
df['title'] = df['title'].str.replace(' - AZFreight', '').str.strip()
```

```python
df
```

```python
cleaned_df = df.dropna(thresh=len(df) * 0.1, axis='columns')
```

```python
cleaned_df = cleaned_df.drop(to_drop, axis='columns')
```

```python
for col in cleaned_df.columns:
    cleaned_df[col] = cleaned_df[col].astype(str).str.strip()
```

```python
cleaned_df = cleaned_df.replace('nan', np.nan)
```
