```python
import pandas as pd
import json
```

```python
with open('./data.jsonl', 'r', encoding='utf-8') as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.json_normalize(data)
```

```python
pd.set_option('display.max_columns', 50)
```

```python
print(df.shape)
df.head()
```

```python
df.to_excel('data.xlsx')
```
