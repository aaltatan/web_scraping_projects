```python
import pandas as pd
import numpy as np
import json
import re
```

```python
with open('devices.jsonl', 'r', encoding='utf-8') as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.json_normalize(data, sep='')
```

```python
pd.set_option('display.max_columns', 100)
```

```python
df['company'] = df['title'].str.split(' ', expand=True)[0]
```

```python
df['year'] = df['informationLaunchAnnounced'].str.extract(r'(\d{4})')
```

```python
df[['heightMm', 'widthMm', 'thicknessMm']] = (
    df['informationBodyDimensions']
    .str
    .extract(r'(\d+\.?\d?)\sx\s(\d+\.?\d?)\sx\s(\d+\.?\d?)', expand=True)
)
```

```python
df['weightG'] = df['informationBodyWeight'].str.extract(r'(\d+(?=\sg))')
```

```python
df[['displayHeightPx', 'displayWidthPx']] = (
    df['informationDisplayResolution']
    .str
    .extract(r'(\d+)\sx\s(\d+)', expand=True)
)
```

```python
df['batteryCapacityMah'] = (
    df['informationBatteryType']
    .str
    .extract(r'(\d+).+mAh')
)
```

```python
df['informationMiscPrice'] = (
    df['informationMiscPrice']
    .str
    .extract(r'(\d+)')
)
```

```python
df.columns = [col.replace(' ', '') for col in df.columns]
```
