```python
import pandas as pd
import numpy as np
import json
```

```python
with open("data.jsonl", "r", encoding="utf-8") as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.json_normalize(data)
```

```python
pd.set_option("display.max_columns", 50)
```

```python
df = df.dropna(how="all", axis="columns")
```

```python
df = df[[
    'nameEng',
    'hasWhatsapp',
    'primaryPhone',
    'primaryEmail',
    'primaryWebsite',
    'facebookUrl',
    'linkedInUrl',
    'twitterUrl',
    'instagramUrl',
    'poBox',
    'address',
    'categories',
]]
```

```python
df["primaryPhone"] = df["primaryPhone"].str.replace(r"\s", "", regex=True)
```
