```python
import pandas as pd
import numpy as np
import json
```

```python
with open("data.jsonl", "r", encoding="utf-8") as file:
    data = (json.loads(line[:-1]) for line in file.readlines())
```

```python
df = pd.json_normalize(data)
```

```python
df = df[
    [
        "companyName",
        "bio",
        "url",
        "email",
        "telephone",
        "fax",
        "career",
        "country_title",
        "degree_title",
        "category_title",
    ]
]
```

```python
df = df.drop_duplicates()
```

```python
df = df.loc[(df["email"] != "") | (df["telephone"] != "")]
```

```python
df = df.sort_values(["country_title", "category_title"])
```

```python
pivot = df.pivot_table(
    index="category_title",
    columns="country_title",
    aggfunc="count",
    fill_value=0,
    margins=True,
)["bio"]
```
